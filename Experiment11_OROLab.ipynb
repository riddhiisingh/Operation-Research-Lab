Q1) A computer center has three expert programmers. The center wants three application programs to be developed. The head of the computer center, after studying carefully the programmers to be deployed, estimates the computer time in minutes required by the experts for the application programmers as follows:

;ab11.png

```python
import numpy as np
from scipy.optimize import linear_sum_assignment
```

```python
cost_matrix = np.array([[120, 100, 80], [80, 90, 110], [110, 140, 120]])
```

```python
row_ind, col_ind = linear_sum_assignment(cost_matrix)
```

```python
optimal = col_ind
```

```python
tc = cost_matrix[row_ind, col_ind].sum()
```

```python
print('Total assignment cost %d' %tc)
```
