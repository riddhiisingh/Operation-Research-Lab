```python
from scipy.optimize import linprog
```

```python
obj = [1, 1]
```

```python
lhs_ineq = [[ 1,  2],[2,  1]]
rhs_ineq = [6, 6]
```

```python
bnd = [(0, float("inf")), (0, float("inf"))]  # Bounds of x & y
```

```python
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
           bounds=bnd,
            method="revised simplex")
```
